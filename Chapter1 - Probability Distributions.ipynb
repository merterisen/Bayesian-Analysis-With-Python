## Continuous distributions

**Normal (Gaussian) Distribution**

The Normal distribution is a continuous probability distribution widely used in statistics and machine learning.  
It is controlled by two parameters:
- Mean (μ): center of the distribution
- Standard deviation (σ): spread of the distribution

*Example:*  
Let's say we have a sample of heights with Mean of 175cm and Standart deviation of 10.  
We can model the heights with normal distribution as:

```python
mu = 175
std = 10

x = np.linspace(100, 250,  100) # lineer 100 number from 100 to 250
y = stats.norm.pdf(x, mu, std)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.plot(x, y)

plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

**Uniform Distribution**

The Uniform distribution is the simplest continuous distribution. All values from start to end are equally likely.  
It is defined by two parameters:
- loc: starting point
- scale: width of the interval
  

*Example:*  
Let's say there is a spinning wheel (degrees from 0 to 360). I will stop it at some random point.  
The pointer is equally likely to stop at any degree on the wheel.

```python
loc = 0
scale = 360

x = np.linspace(-30, 390, 100)
y = stats.uniform.pdf(x, loc, scale)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.plot(x, y)

plt.xlabel('Angle (degrees)')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

**Beta Distribution**

The Beta distribution is a continuous probability distribution defined over the interval [0,1]. It is used to model uncertainty about probabilities.  
It is controlled by two parameters:
- α: number of successes + 1
- β: number of failures + 1

*Example:*  
Suppose a sports analyst trying to estimate how likely a player is to score a penalty.  
So far, the player has taken 10 shoot and scored 7 times.

```python
alpha = 7 + 1 # number of successes + 1
beta = 3 + 1 # number of failures + 1

x = np.linspace(0, 1, 100)
y = stats.beta.pdf(x, alpha, beta)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.plot(x, y)

plt.xlabel('Scoring Probability')
plt.ylabel('Density')
plt.title(f'Beta Distribution: α={alpha}, β={beta}')
plt.grid(alpha=0.2)
plt.show()
```

**Gamma Distribution**

The Gamma distribution is a continuous probability distribution defined over positive real numbers (x > 0).  
It is controlled by two parameters:
- α (shape): number of events
- β (rate) or θ=1/β (scale): time between events

*Example:*  
Suppose buses arrive at a station randomly at a constant average rate of 1 bus per 10 minutes.  
I want to know the distribution of time until the 3rd bus arrives.  

The peak shows the most likely time for the 3rd bus to arrive.

```python
alpha = 3 # we are waiting for 3 buses
scale = 10 # average wait per bus is 10 minutes

x = np.linspace(0, 40, 100)
y = stats.gamma.pdf(x, alpha, scale=scale) # you have to write it like scale=scale, otherwise plots wrong I dont know why


# Plot Settings
plt.figure(figsize=(3, 2))
plt.plot(x, y)

plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.title(f'Gamma Distribution: α={alpha}, scale={scale}')
plt.grid(alpha=0.2)
plt.show()
```

## Discrete distributions

**Bernoulli Distribution**

The Bernoulli distribution models a random variable that takes the value: 
- 1 with probability p
- 0 with probability 1 - p  

Only one trial is involved in Bernoulli Distribution.  

*Example:*  
Let’s say flip a fair coin (1 = heads, 0 = tails) then probability p is 0.5   



```python
p = 0.5

x = [0, 1] # Possible outcomes
y = stats.bernoulli.pmf(x, p)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.bar(x, y, width=0.3)

plt.ylim(0, 1)
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p = {p})')
plt.grid(alpha=0.2)
plt.show()
```

**Binomial Distribution**

The Binomial distribution models the number of successes in n independent Bernoulli trials with success probability p.  
(Remember, in Bernoulli only one trial was involved.)  

- n: number of trials
- p: probability of success in each trial   
- x: 0 to n

*Example:*  
Imagine flipping a fair coin 10 times. The probability of heads is p=0.5  
The most likely number of heads is around n/2, which is expected since p = 0.5



```python
n = 10
p = 0.5

x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.bar(x, y, width=0.3)

plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution: n={n}, p={p}')
plt.grid(alpha=0.2)
plt.show()
```

**Poisson Distribution**

The Poisson distribution models the number of events occurring in a fixed interval of time or space, assuming: 

Events happen independently  
At a constant average rate λ  
No two events happen simultaneously  

- λ: expected number of events per interval
- k: number of events (0, 1, 2, …)

*Example:*  
Suppose receiving emails at an average rate of 4 per hour.  
The probability of receiving exactly k emails in one hour:



```python
λ = 4

x = np.arange(0, 15)
y = stats.poisson.pmf(x, λ)


# Plot Settings
plt.figure(figsize=(3, 2))
plt.bar(x, y, width=0.3)

plt.xlabel('k Number of Emails')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution: λ = {λ}')
plt.grid(alpha=0.2)
plt.show()
```
